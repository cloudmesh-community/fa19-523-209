```python
#Split the dataset into input data x and target value y
x,y = da.drop('price', axis = 1), da.price
```

```python
#Import linear models lib, data test/train splitting lib and errors lib
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
#divide both x and y into 80% train and 20% test using train_test_split and view their dimentions
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=13) #train_size = 0.8
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
#Import three linear regression models and two ensemble models
from sklearn.linear_model import LinearRegression as LR, ElasticNet as ENet, Ridge as R, Lasso as L
from sklearn.ensemble import ExtraTreesRegressor as ETR, GradientBoostingRegressor as GBR
```

```python
#Names conventions
algorithms = [LR(),ENet(),R(),L(),ETR(),GBR()]
names=['LinearRegression','ElasticNet', 'Ridge', 'Lasso', 'ExtraTreesRegressor', 'GradientBoostingRegressor']
```

```python
#Function to build and evaluate 6 different models
def zmetrics(x,y,algorithms=algorithms,names=names):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
    r2=[]
    rmse=[]
    mae=[]
    for i in range(len(algorithms)):
        algorithms[i].fit(x_train,y_train)
        prediction=algorithms[i].predict(x_test)
        r2.append(r2_score(y_test,prediction))
        rmse.append(mean_squared_error(y_test,prediction)**0.5)
        mae.append(mean_absolute_error(y_test,prediction))
    metrics=pd.DataFrame(columns=['R2_Score','Root Mean Square Error', 'Mean Absolute Error'],index=names)
    metrics['R2_Score']=r2
    metrics['Root Mean Square Error']=rmse
    metrics['Mean Absolute Error']=mae
    return metrics.sort_values('R2_Score',ascending=False)
        
```

```python
#Results of running zmetrics on selected subset of house point-of-sale data
zmetrics(x,y)
```
